Lung Cancer Dataset https://www.kaggle.com/datasets/akashnath29/lung-cancer-dataset

```python
import pandas as pd

#
df = pd.read_csv('dataset.csv')
```

```python
import matplotlib.pyplot as plt

#VISUALIZATION

# target (LUNG_CANCER)
lung_cancer_counts = df['LUNG_CANCER'].value_counts()

# pie chart
plt.figure(figsize=(11, 8))
plt.pie(lung_cancer_counts, labels=lung_cancer_counts.index, autopct='%1.1f%%', startangle=90, colors=['red', 'blue'])
plt.title('Distribution of Lung Cancer Cases')
plt.show()

```

```python
# This is to show the dataset of different genders
gender_counts = df['GENDER'].value_counts()

# Barchart for gender
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Distribution of GENDER')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
