```python
import numpy as np
array_2x2 = np.array([[1, 2],[3, 4]])
print("1. Numpy array of size 2x2:\n", array_2x2)
```

```python
import numpy as np
random_3x4 = np.random.rand(3, 4)
print("\n2. Random numpy array of size 3x4:\n", random_3x4)
```

```python
import numpy as np
ones_array = np.ones((2, 3))
zeroes_array = np.zeros((3, 2))
print("\n3.Numpy array of ones (2x3):\n", ones_array, "\n")
print("  Numpy array of zeroes (3x2):\n", zeroes_array)
```

```python
import numpy as np
random_6x3 = np.random.rand(6, 3)
transpose_6x3 = random_6x3.T
print("\n4.Random numpy array of size 6x3:\n", random_6x3, "\n")
print("  Transposed array (3x6):\n", transpose_6x3)
```
